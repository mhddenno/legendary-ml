Assumptions of Linear Regression to make it work

1. Linearity
(Linear relationship between Y and each X)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(0) # for reproducing the same result
X_linear = 2 * np.random.rand(100, 1)
y_linear = 3 + 4 * X_linear + np.random.randn(100, 1)

linear_reg = LinearRegression()
linear_reg.fit(X_linear, y_linear)

plt.scatter(X_linear, y_linear, color='blue')
plt.plot(X_linear, linear_reg.predict(X_linear), color='red', linewidth=3)
plt.title('Example of Linearity')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

################################ Not Linear
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

np.random.seed(0)
X_non_linear = 2 * np.random.rand(100, 1)
y_non_linear = 1 + 2 * X_non_linear + 15 * X_non_linear**2 + np.random.randn(100, 1)

# Fit a polynomial regression model
degree = 2
poly_reg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly_reg.fit(X_non_linear, y_non_linear)

# Plot the data and the polynomial regression curve
plt.scatter(X_non_linear, y_non_linear, color='blue')
plt.plot(X_non_linear, poly_reg.predict(X_non_linear), color='red', linewidth=1, linestyle= 'dotted')
plt.title('Example of Non-Linearity')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

2. Homoscedasticity
(Equal variance)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate homoscedastic data
np.random.seed(0)
X_homoscedastic = 2 * np.random.rand(100, 1)
y_homoscedastic = 3 + 4 * X_homoscedastic + np.random.randn(100, 1)

# Fit a linear regression model
linear_reg_homoscedastic = LinearRegression()
linear_reg_homoscedastic.fit(X_homoscedastic, y_homoscedastic)

# Plot the data and the linear regression line
plt.scatter(X_homoscedastic, y_homoscedastic, color='blue')
plt.plot(X_homoscedastic, linear_reg_homoscedastic.predict(X_homoscedastic), color='red', linewidth=3)
plt.title('Example of Homoscedasticity')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Generate heteroscedastic data
np.random.seed(0)
X_heteroscedastic = 2 * np.random.rand(100, 1)
y_heteroscedastic = 3 + 4 * X_heteroscedastic + X_heteroscedastic*np.random.randn(100, 1)

# Fit a linear regression model
linear_reg_heteroscedastic = LinearRegression()
linear_reg_heteroscedastic.fit(X_heteroscedastic, y_heteroscedastic)

# Plot the data and the linear regression line
plt.scatter(X_heteroscedastic, y_heteroscedastic, color='blue')
plt.plot(X_heteroscedastic, linear_reg_heteroscedastic.predict(X_heteroscedastic), color='red', linewidth=3)
plt.title('Example of Heteroscedasticity')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
