Understading the Mean Value

```python
#!/usr/bin/env python

# Mean
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) # it means that every time you run the code, you will get the same set of random numbers 
data_points = np.random.randn(30)
print("the data:",data_points)
mean_value = np.mean(data_points)
print("the mean:",mean_value)
# Visual
plt.scatter(range(1, len(data_points) + 1), data_points, label='Data Points')
plt.axhline(mean_value,color='red', linestyle='dashed', linewidth='2', label=f'Mean: {mean_value:.2f}')
plt.xlabel('Data Point Index')
plt.ylabel('Value')
plt.title('Scatter Plot with Mean')
plt.legend()
plt.show()
```

Mean in 2D

```python
#!/usr/bin/env python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data_points = np.random.randn(30,2)
mean_values = np.mean(data_points,axis=0) # axis=0: Calculate the mean along the rows (along each column). axis=1: Calculate the mean along the columns (along each row).
print(mean_values)
plt.scatter(data_points[:,0],data_points[:,1],label='Data Points')
plt.scatter(mean_values[0], mean_values[1], color='red', marker = 'x', s=100, label=f'Mean: {mean_values}')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Scatter Plot with Mean (2D)')
plt.legend()

plt.show()
```
